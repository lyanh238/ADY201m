## Advanced Operations (Requirements)

#### Calculate the total number of students in each grade level

```python
%%sql
select "Elementary_Middle_or_High_School", sum("COLLEGE_ENROLLMENT") as "Total_Number_Students"
from ChicagoPublicSchools
group by "Elementary_Middle_or_High_School";
```

#### Find the top 5 student groups with the largest number and highest safety


```python
%%sql
select "COMMUNITY_AREA_NAME", sum("COLLEGE_ENROLLMENT") as "Total_Number_Students", 
        avg("SAFETY_SCORE") as "Safety_Score_each_Area_group"
from ChicagoPublicSchools
Group By "COMMUNITY_AREA_NAME"
Order By "Total_Number_Students" DESC, "Safety_Score_each_Area_group" DESC  LIMIT 5;
```

```python
%%sql
SELECT Name_of_school, SUM(COLLEGE_ENROLLMENT) AS Total_Enrollment, Safety_Score
FROM ChicagoPublicSchools
GROUP BY Name_of_school
ORDER BY Total_Enrollment DESC, Safety_Score DESC
LIMIT 5;
```

#### Calculate average of AVERAGE_STUDENT_ATTENDANCE of 10 schools with least SAFETY_SCORE

```python
%%sql
SELECT avg(cast(replace(Average_Student_Attendance,'%','') as real)) as "Average Student Attendance"
FROM (SELECT Average_Student_Attendance, Safety_Score
      FROM ChicagoPublicSchools
      Where Safety_Score != 'None'
      ORDER BY Safety_Score
      LIMIT 10)
```

#### Find 10 High school with highest College_Eligibility_score.

```python
%%sql
Select "NAME_OF_SCHOOL", "College_Eligibility__"
From ChicagoPublicSchools
Where ("Elementary_Middle_or_High_School" = 'HS') and ("College_Eligibility__" != 'NDA') 
Order By Cast("College_Eligibility__" as real) DESC Limit 10;
```

```python
%%sql
Select count("College_Eligibility__") as "NDA College_Eligibility"
From ChicagoPublicSchools
Where ("Elementary_Middle_or_High_School" = 'HS') and ("College_Eligibility__" == 'NDA') 
```

```python
%%sql
Select count(*) as "Total High School"
From ChicagoPublicSchools
Where ("Elementary_Middle_or_High_School" = 'HS')
```

### Draw positions of all schools in Chicago

#### Using Folium

```python
!pip3 install folium==0.5.0
import folium

print('Folium installed and imported!')
```

```python
LocationSchool = Latitude_Longitude.DataFrame()
LocationSchool.head()
```

```python
LocationSchool.shape
```

#### Some schools have special characters in there name - modify it first

```python
#Save latitudes, longitudes and name of school
latitudes = list(LocationSchool.Latitude)
longitudes = list(LocationSchool.Longitude)
labels = list(LocationSchool.NAME_OF_SCHOOL)
olabels = list(LocationSchool.NAME_OF_SCHOOL)

#Find, check and replace special characters into ' '
special_characters = """"'"""

def checkSpecial(st):
    for i in range(len(st)):
        if st[i] in special_characters:
            return True
    return False
 
def replaceSpecial(st):
    new = st
    for i in range(len(st)):
        if st[i] in special_characters:
            new = st.replace(st[i],' ')
    return new

index_of_special_name = []
for i in range(200,350):
    if (checkSpecial(labels[i])):
        index_of_special_name.append(i)
        labels[i] = replaceSpecial(labels[i])
        
```

### Draw positions into map

```python
LocationS1 = folium.Map(location=[41.86, -87.6], zoom_start=11)

schools = folium.map.FeatureGroup()

for lat, lng, in zip(latitudes, longitudes):
    schools.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

for i in range(0,len(labels)):
    folium.Marker(location = [latitudes[i],longitudes[i]], popup = labels[i]).add_to(LocationS1)

#fg.add_to(LocationS)
#LocationS.save("Chicago.html")

LocationS1.add_child(schools)
```

### Group the marked points into a cluster

```python
from folium import plugins
LocationS1 = folium.Map(location=[41.86, -87.6], zoom_start=11)

schools = folium.map.FeatureGroup()


marker_cluster = plugins.MarkerCluster().add_to(LocationS1)
    
for i in range(0,566):
    folium.Marker(location = [latitudes[i],longitudes[i]], popup = labels[i]).add_to(marker_cluster)

#fg.add_to(LocationS)
#LocationS.save("Chicago.html")

LocationS1.add_child(schools)
```

#### Now here are some special school name

```python
spLat = []
spLong = []
spLabel = []
for i in index_of_special_name:
    spLat.append(latitudes[i])
    spLong.append(longitudes[i])
    spLabel.append(olabels[i])
    
import pandas as pd

data = {'latitude': spLat, 'longitude': spLong, 'label': spLabel}
schoolname = pd.DataFrame(data)
```
