```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.edge.service import Service
import os
import pandas as pd
import time

# Function to scrape data from each page
def scrape_something(driver):
    html = driver.page_source  # Get current page source
    soup = BeautifulSoup(html, 'html.parser')

    property_cards = soup.find_all("div", class_="re__card-info-content")

    land_list = []
    # Lặp từng cái thẻ và đưa ra thông tin
    for card in property_cards:
        try:
            location = card.find("div", class_="re__card-location").text.strip()
        except AttributeError:
            location = "N/A"
        
        try:
            price = card.find("span", class_="re__card-config-price").text.strip()
        except AttributeError:
            price = "N/A"
        
        try:
            bedrooms = card.find("span", class_="re__card-config-bedroom").text.strip()
        except AttributeError:
            bedrooms = "N/A"
        
        try:
            bathrooms = card.find("span", class_="re__card-config-toilet").text.strip()
        except AttributeError:
            bathrooms = "N/A"
        
        try:
            area = card.find("span", class_="re__card-config-area").text.strip()
        except AttributeError:
            area = "N/A"
        
        try:
            title = card.find("span", class_="pr-title").text.strip()
        except AttributeError:
            title = "N/A"

        # Skip entries with "Giá thỏa thuận"
        if "Giá thỏa thuận" not in price:
            land_list.append({
                "Title": title,
                "Location": location,
                "Price": price.replace("₫", "").replace(".", "").strip(),
                "Area": area,
                "Bedrooms": bedrooms,
                "Bathrooms": bathrooms
            })
    
    return land_list

# Function to save data to CSV
def save_data(land_data):
    # Convert the scraped data into a DataFrame
    df = pd.DataFrame(land_data, columns=['Title', 'Location', 'Price', 'Area', 'Bedrooms', 'Bathrooms'])

    # Check if the file exists, and append or create a new file
    if not os.path.isfile('land_price.csv'):
        df.to_csv('land_price.csv', mode='w', header=True, index=False)
    else:
        df.to_csv('land_price.csv', mode='a', header=False, index=False)

# Setup WebDriver for Edge
service = Service(executable_path=r'C:\\Users\\Dell\\Downloads\\edgedriver_win64\\msedgedriver.exe')
driver = webdriver.Edge(service=service)

# Loop through pages i to n and scrape data in one window
for i in range(681, 1000):
    url = f'https://batdongsan.com.vn/nha-dat-ban/p{i}'
    driver.get(url)  # Navigate to the page
    time.sleep(2)  # Allow time for page to load
    
    # Scrape data from the current page
    land_data = scrape_something(driver)
    
    # Save data to CSV after each page
    if land_data:  # Save only if there is data
        save_data(land_data)
        print(f"Page {i}: Scraped {len(land_data)} properties")
    else:
        print(f"Page {i}: No properties found")

driver.quit()  # Close the browser window when done
print("Scraping completed.")
```
