```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
csv_file_path = "individual_stocks_5yr/individual_stocks_5yr/AAL_data.csv" 

stock_data = pd.read_csv(csv_file_path)

print(stock_data.head())

stock_data.isnull().sum()

stock_data['Date'] = pd.to_datetime(stock_data['Date'])



stock_data.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume','Name']
```

```python
X = stock_data[['Open', 'High', 'Low', 'Volume']]
y = stock_data['Close']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

model = LinearRegression().fit(X_train, y_train)
predictions = model.predict(X_test)
print(predictions)

plt.figure(figsize=(10, 5))
plt.scatter(y_test, predictions, color='blue', label='Predicted vs Actual')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', lw=2, label='Ideal Fit')
plt.xlabel('Actual Close Price')
plt.ylabel('Predicted Close Price')
plt.title('Linear Regression: Actual vs Predicted Close Prices')
plt.legend()
plt.show()
```
